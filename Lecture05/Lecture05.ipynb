**Many to one:** Los datos de entrada son una secuencia, pero la salida es un vector de tamaño fijo, no una secuencia.

Ejemplo: Análisis de sentimiento, la entrada es un texto y la salida es una etiqueta de clase.

# **Diferentes tipos de tareas de modelado de secuencias**

<img src="Figs/RNN13.PNG" style="max-height: 500px; width: auto;">

**One to many:** Los datos de entrada están en un formato estándar (no una secuencia), la salida es una secuencia.

Ejemplo: Subtítulos de imágenes, donde la entrada es una imagen, la salida es una descripción de texto de esa imagen


# **Diferentes tipos de tareas de modelado de secuencias**

**Many to many:** Tanto las entradas como las salidas son secuencias. Puede ser directo o retrasado.

Ejemplo: Subtítulos de video, es decir, describir una secuencia de imágenes a través de texto (directo).

# **Soluciones a los problemas de gradiente de desaparición/explosión**

- Gadient clipping: Establecer un valor máximo para los gradientes si crecen demasiado (solo resuelve el problema del gradiente explosivo)
- Propagación hacia atrás truncada a través del tiempo (TBPTT): Simplemente limita el número de pasos de tiempo que la señal puede propagarse hacia atrás después de cada pasada hacia adelante. Por ejemplo, incluso si la secuencia tiene 100 elementos/pasos, solo podemos propagar hacia atrás hasta 20 más o menos
- Memoria a largo/corto plazo (LSTM): Utiliza una celda de memoria para modelar dependencias de largo alcance y evitar problemas de gradiente que desaparecen.


Hochreiter, Sepp, and Jürgen Schmidhuber. "Long short-term memory." Neural computation 9, no. 8 (1997): 1735-1780.
