# Defining Efficient Data Loaders



How to set up a `DataLoader` for a folder containing image files.

# Obtaining the Dataset

- This downloads the datasets png files:

```python
!git clone "https://github.com/rasbt/mnist-pngs"
```

```python
import os
import pandas as pd

base_path = "mnist-pngs"

def build_csv(split):
    rows = []
    split_path = os.path.join(base_path, split)

    for label in sorted(os.listdir(split_path), key=int):
        label_path = os.path.join(split_path, label)

        if not os.path.isdir(label_path):
            continue

        for img in sorted(os.listdir(label_path)):
            if img.endswith(".png"):
                rows.append({
                    "file_path": os.path.join(split, label, img),
                    "label": int(label)
                })

    return pd.DataFrame(rows)

df_train = build_csv("train")
df_test  = build_csv("test")
```

```python
df_train.to_csv(os.path.join(base_path,"train.csv"), index=False)
df_test.to_csv(os.path.join(base_path,"test.csv"), index=False)
```

## Creating a validation split

- MNIST doesn't come with a validation set partition, so we are creating it here from the training set, using 10% of the training data for validation.

```python
df_train = df_train.sample(frac=1, random_state=123)

loc = round(df_train.shape[0]*0.9)
df_new_train = df_train.iloc[:loc]
df_new_val = df_train.iloc[loc:]

df_new_train.to_csv('mnist-pngs/new_train.csv', index=None)
df_new_val.to_csv('mnist-pngs/new_val.csv', index=None)
```
